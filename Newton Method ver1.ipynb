```python
import numpy as np
from matplotlib import pyplot as plt

def f(x): 
    return x**3 - 3 * x**2 + x
def fprime(x) :
    return 3*x**2-6*x+1

x = np.linspace(-1, 3, 4000)
y = f(x)

plt.plot(x, y)
plt.plot(0, f(0), 'ro')
plt.plot(x, fprime(0)*(x-0)+f(0), 'r:')
plt.plot(1, f(1), 'go')
plt.plot(x, fprime(1)*(x-1)+f(1), 'g--')
plt.plot(-(f(1)/fprime(1))+1, f(-(f(1)/fprime(1))+1), 'bo')
plt.plot(x, fprime(-(f(1)/fprime(1))+1)*(x+(f(1)/fprime(1))-1)+f(-(f(1)/fprime(1))+1), 'b--')

plt.xlim(-3.5, 5.5)
plt.ylim(-4, 2)
plt.xticks(np.arange(-3, 6))
plt.yticks(np.arange(-4, 2))

ax = plt.gca()
ax.spines['bottom'].set_position(('data',0))
ax.spines['left'].set_position(('data',0))

plt.plot([1,1],[0,f(1)], color ='black', linewidth=0.5, linestyle="--")

plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("title")
plt.show()
        
```
