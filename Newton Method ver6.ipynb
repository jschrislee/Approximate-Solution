```python
import numpy as np
from matplotlib import pyplot as plt

def f(x): 
    return x**3 - 3 * x**2 + x
def fprime(x) :
    return 3*x**2-6*x+1
def xintercept(x) :
    return -(f(x)/fprime(x))+x
def tangent(x, t) :
    return fprime(t)*(x-t)+f(t)

n=int(input('반복 횟수 : '))
s=float(input('x0 = '))

x = np.linspace(-1, 3, 4000)
y = f(x)
plt.plot(x, y)

ax = plt.gca()
ax.spines['bottom'].set_position(('data',0))
ax.spines['left'].set_position(('data',0))

for i in range(1,n) :
    if i==1 : 
        plt.plot(s, f(s), 'ro')
        plt.plot(x, tangent(x, s), 'r--')
        print('xintercept of x0 = ', xintercept(s))
        t=xintercept(s)
    
    print(xintercept(t))
    plt.plot(t, f(t), 'ro')
    plt.plot(x, tangent(x, t), 'r--')
    print('xintercept of x= ', t, ':', xintercept(t))
    t=xintercept(t)
```
